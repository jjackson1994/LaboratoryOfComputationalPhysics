4\. load the binary file named *credit_card.dat* and convert the data into the real credit-card number.
Each line correspond to a credit card number.
Each character is composed by 6 bit (even the space) and the last 4 bit are just a padding

**hint**: use the `chr()` function to convert a number to a char

```python
f = open('credit_card.dat', 'r')
c_a =[] #card array
for line in f:
    c_a.append(line)
f.close()

def extract_num(c_a):

    num_array = []
    for row in c_a:
        row_trim=row[:-5]
        card_num = ''
        for i in range(0,int(len(row_trim)/6)):
            card_bin=row_trim[6*i:6*(i+1)]
            card_bin=int(row_trim[6*i:6*(i+1)],2)
            card_num+=chr(card_bin)
        num_array.append(card_num)
    return num_array# lines

print(extract_num(c_a))
```
