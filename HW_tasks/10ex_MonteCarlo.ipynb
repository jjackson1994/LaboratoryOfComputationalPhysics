1\. **Radioactive decay chain**

${\rm Tl}^{208}$ decays to ${\rm Pb}^{208}$ with a half-lieve of 3.052 minutes. Suppose to start with a sample of 1000 Thallium atoms and 0 of Lead atoms.

* Take steps in time of 1 second and at each time-step decide whether each Tl atom has decayed or not, accordingly to the probability $p(t)=1-2^{-t/\tau}$. Subtract the total number of Tl atoms that decayed at each step from the Tl sample and add them to the Lead one. Plot the evolution of the two sets as a function of time  

```python
import numpy as np
from numpy import log as ln
import matplotlib.pyplot as plt
import math 

half_life= 3.052*60 #s
N_Tl=[1000]
N_Pb=[0]
times=[0]
print(half_life)
def p(t,tau):
    return 1-2**(-t/tau)

def p_2(t,tau):
    return (ln(2)/tau)*2**(-t/tau)

T = 60*8 #8 mins

survival_prob = np.random.rand(N_Tl[-1]) #note this is only generated once made the mistake of dooing thiseach iternation and it fell too fast  
for t in range(1, T):
    prob = p(t, half_life) #probability that an atoms will have decayed by time t 
    #decay_trial = np.random.rand(N_Tl[-1]) # random dist size of atoms left note this does the subraction
    Tl_remaining = survival_prob[prob < survival_prob] #mask out decayed atoms
    N_Tl.append(len(Tl_remaining))
    N_Pb.append(N_Tl[0]-len(Tl_remaining)) 
    times.append(t)

plt.figure(figsize=(10,6))
plt.plot(times, N_Pb,label ='Lead')
plt.plot(times, N_Tl, label ='Thallium')
plt.xlabel("time / s")
plt.ylabel("Number of atoms") 
plt.legend() 
plt.show()
```

Make a plot showing the number of atoms that have not decayed as a function of time

$$p(t) = 2^{-\frac{t}{\tau}} \frac{\ln 2}{\tau}$$
Pre normalised as
$$\int_{0}^{+\infty} p(t) dt = \int_{0}^{+\infty} 2^{-\frac{t}{\tau}} \frac{\ln 2}{\tau} dt = 1$$
Following inverse transform method from the lecture 
$$\int_{0}^{t(z)} p(t') dt' = \int_{0}^{t(z)} 2^{-\frac{t'}{\tau}} \frac{\ln 2}{\tau} dt' = 1-2^{-\frac{t(z)}{\tau}} = z$$
Therefore inverse is
$$t(z) = -\frac{\tau}{\ln 2}  \ln{(1-z)}$$

```python
def expon_pdf(t, mu=1, tau =  3.052*60 ):
    """PDF of exponential distribution."""
    return (ln(2)/tau)*2**(-t/tau)

def expon_cdf(t, mu=1, tau = 3.052*60):
    """CDF of exponetial distribution."""
    return 1 -2**(-t/tau)

def expon_icdf(z, tau = 3.052*60):
    """Inverse CDF of exponential distribution - i.e. quantile function."""
    return (-tau/ln(2))*ln(1-z)
```

2\. **Rutherford Scattering**

The scattering angle $\theta$ of $\alpha$ particles hitting a positively charged nucleus of a Gold atom ($Z=79$) follows the rule:

$$
\tan{\frac{1}{2} \theta} = \frac{Z e^2} {2\pi \epsilon_0 E b}
$$

where $E=7.7$ MeV and $b$ beam is the impact parameter. The beam is represented by a 2D gaussian distribution with $\sigma=a_0/100$ for both coordinates ($a_0$ being the Bohr radius). Assume 1 million $\alpha$ particles are shot on the gold atom.

Computing the fraction of particles that "bounce back",i.e. those particle whose scattering angle is greater than $\pi/2$ (which set a condition on the impact parameter $b$)

```python
def theta(b):
    e = 1.6*10**(-19)
    E = 7.7*(10**6)*e #MeV
    Z = 79 #As gold
    a0 = 5.29*(10**-11)
    eps0 = 8.854*(10**-12)
    return 2*np.tan((Z*e**2)/(2*np.pi*eps0*E*b))

a0 = 5.29*(10**-11)
sigma = a0/100

N= 1000
b = np.linspace(-sigma, sigma, N) 
theta_val = theta(b)
#mask=(theta_val < np.pi/2) | (np.pi < theta_val/2)
mask=(np.pi < theta_val/2)
bounce_back = theta_val[mask]
print('Back scatter fraction = ', len(bounce_back)/N)
plt.figure(figsize=(8, 6))
plt.plot(b, theta_val)
plt.xlabel('b')
plt.ylabel('theta')                    
```

3\. **Monte Carlo integration: hit/miss vs mean value method**

Consider the function 

$$f(x) =\sin^2{\frac{1}{x(2-x)}}$$

* Compute the integral of $f(x)$ between 0 and 2 with the hit/miss method. Evaluate the error of your estimate
* Repeat the integral with the mean value method. Evaluate the error and compare it with the previous one

```python
def f(x):
    return np.square(np.sin(1/(x*(2-x))))

def hit_miss(funct,n, x_low, x_high):
    x_ = np.random.uniform(low=x_low, high=x_high, size=n)
    y_ = np.random.uniform(low=0, high=1, size=n)
    hit_mask = (y_ <= funct(x_))
    miss_mask = (y_>funct(x_))
    hits = [x_[hit_mask],y_[hit_mask]]
    miss = [x_[miss_mask], y_[miss_mask]]
    return [hits,miss]

x_low = 0
x_high = 2 
x = np.linspace(x_low, x_high, 100) 
y = f(x)

n = 1000
hm=hit_miss(f,n,x_low,x_high)
hit=len(hm[0][0])
miss = len(hm[1][0])
I = n/hit
print(hit,miss)
print(I)
plt.figure(figsize=(8, 6))
plt.plot(x, y,label ='test')
plt.scatter(hm[0][0], hm[0][1],label ='hits')
plt.scatter(hm[1][0], hm[1][1],label ='misses')
```

Mean Value Method
$$I=\frac{b-a}{N} \sum_{i=1}^{N} f(x_i)$$

Where the standard devation is given by
$$\sigma = V\frac{\sqrt{{\rm var}\, f}}{\sqrt{N}}$$

```python
def mean_val(f, x_low, x_high, n):
    x_ = np.random.uniform(low=x_low, high=x_high, size=n)
    y_ = f(x_)
    var= np.var(y_)
    return [np.sum(y_)*(x_high - x_low)/n, (x_high-x_low)*np.sqrt(np.var(y_)/n)]

print(mean_val(f, x_low, x_high, n))
```

4\. **Monte Carlo integration in high dimension**

* Start of by computing the area of a circle of unit radius, by integrating the function 

$$
f(x,y)=
\left\{
\begin{array}{ll}
      1 & x^2+y^2\le 1 \\
      0 & {\rm elsewhere}
\end{array} 
\right.
$$

* Generalize the result for a 10D sphere



```python
n=110
def f_c(x,y):
    Rsq = x**2 +y**2
    Rsq[(Rsq<=0) | (Rsq>1)] = 0
    Rsq[(0<Rsq) & (Rsq<1)] =1
    return Rsq

x_c = np.linspace(-1,1,n)
y_c = np.linspace(-1,1,n) 
x_1, y_1 = np.meshgrid(x_c, y_c)
f_xy= f_c(x_1,y_1) 

hit=len(f_xy[f_xy==1])
miss=len(f_xy[f_xy==0])
print('hit', hit)
print('miss', miss)
area = hit/miss
print(area)
plt.figure(figsize = (8,8))
plt.imshow(f_xy, extent=[-1, 1, -1, 1], aspect='auto')
```
